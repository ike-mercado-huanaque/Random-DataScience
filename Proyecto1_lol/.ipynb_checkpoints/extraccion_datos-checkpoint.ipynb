```python
def get_player_stats(region,matchid, summoner_id):
    match = lol_watcher.match.by_id(region, matchid)
    for n in range(10):
        if match['info']['participants'][n]['summonerId'] == summoner_id:
            return match['info']['participants'][n]
def get_participants_match(region,matchid):
    match = lol_watcher.match.by_id(region, matchid)
    participants = [match['info']['participants'][n]['summonerId'] for n in range(10)]
    return participants
def get_record_stats(usuario,region,count=10,queue=None,showmatch=False):
    matches = lol_watcher.match.matchlist_by_puuid(region, usuario["puuid"],count=count,queue=queue)
    match_stats = []
    for match_id in matches:
        stats = get_player_stats(region,match_id, usuario["id"])
        #No agregar partidas que terminaron en Remake
        if stats['gameEndedInEarlySurrender'] == False:
            match_stats.append(stats)
        if showmatch:
            match = lol_watcher.match.by_id(region, match_id)
            print('queueId: ',match['info']['queueId'],',')
            print('gameId: ',match['info']['gameId'],',')
            print('gameMode: ',match['info']['gameMode'],',')
            print('gameName: ',match['info']['gameName'],',')
            print('gameType: ',match['info']['gameType'],'\n')
    return match_stats
```

```python
def get_record_stats_by_name(names_match,region,count=10,queue=None,showmatch=False):
    if isinstance(names_match,str):
        users = [names_match]
    elif isinstance(names_match,list):
        users = names_match
    else:
        raise TypeError("Parameter 'names_match' must be list or string")
    all_stats = dict()
    for name in users:
        user_dict = lol_watcher.summoner.by_name(region, name)
        user_stats = get_record_stats(user_dict,region,count,queue,showmatch)
        all_stats[name] = user_stats
    return all_stats
def get_record_stats_by_id(names_id,region,count=10,queue=None,showmatch=False):
    if isinstance(names_id,str):
        users = [names_id]
    elif isinstance(names_id,list):
        users = names_id
    else:
        raise TypeError("Parameter 'names_match' must be list or string")
    all_stats = dict()
    for u_id in users:
        user_dict = lol_watcher.summoner.by_id(region, u_id)
        user_stats = get_record_stats(user_dict,region,count,queue,showmatch)
        all_stats[u_id] = user_stats
    return all_stats
```

```python
def plot_winning_strike(stats,count=10,team=None,label=None):
    print(len(stats))
    x= np.zeros(count+1)
    #Normalizacion para unrankeds
    x[0:min(count,len(stats))] += np.array([int(stats[n]['win']) for n in range(min(count,len(stats)))])
    x = np.flip(x)
    x[1+count-len(stats):] = 2*x[1+count-len(stats):] - 1
    for i in range(count):
        x[i+1] = x[i+1]+x[i]
    if label != None:
        plt.plot(x,'o--',label=label)
        plt.legend(loc='lower left')
    elif label == None:
        plt.plot(x,'o--')
    return None
    
```

```python
def get_previous_matches(summoner_puuid, region, match_id,queue=None,count=10):
    try:
        #ARREGLAR ESTA LINEA PARA QUE BUSQUE EL MATCHID
        previous_matches = []
        found_match = False
        idx_init=0
        search = min([count+20,100])
        keep = True
        while (not found_match) or keep:
            matchlist = lol_watcher.match.matchlist_by_puuid(region, summoner_puuid,count = search,start=idx_init,queue=queue)
            if len(matchlist) == 0:
                break
            elif keep:
                for match in matchlist:
                    if match == match_id:
                        found_match = True
                    elif found_match:
                        previous_matches.append(match)
                        if len(previous_matches) >= count:
                            keep = False
                            break
                idx_init += search
            
        return previous_matches
    except ApiError as err:
        print(f"Error retrieving previous matches: {err}")
        return None
def get_record_stats_by_match(matchid,region,queue=None,count=10,show=True):
    try:
        players = get_participants_match(region=region,matchid=matchid)   
    except:
        print("Error getting players")
        return None
    record_match = dict()
    if show:
        print(" || Player || matches || time executing || \n")
    for p_id in players:
        ts = time.time()
        match_stats = []
        p_puuid = lol_watcher.summoner.by_id(region,p_id)['puuid']
        if count > 0:
            matches = [matchid] + get_previous_matches(summoner_puuid=p_puuid,region=region,match_id=matchid,queue=queue,count=count)
        else:
            matches = [matchid]
        for match_id in matches:
            stats = get_player_stats(region,match_id, p_id)
            #NO AGREGAR PARTIDAS QUE TERMINARON EN REMAKE
            if stats['gameEndedInEarlySurrender'] == False:
                match_stats.append(stats)        
        record_match[p_id] = match_stats
        te = time.time()
        if show:
            print(f"|| {lol_watcher.summoner.by_id(region,p_id)['name']} || {len(matches)} || {round(te-ts,2)} [s] || \n")
    return record_match

def get_match_stats(matchid,region,queue=None,count= 10,include_teams = False):
    if isinstance(matchid,str):
        matchid = [matchid]
    elif isinstance(matchid,list):
        pass
    else:
        raise TypeError("Solo formato string o list son permitidos.")
    matches_dict = dict()
    for m_id in matchid:
        matches_dict[m_id] = lol_watcher.match.by_id(region, m_id)['info']['participants']
        if include_teams:
            matches_dict[m_id].append(lol_watcher.match.by_id(region, m_id)['info']['teams'])
    return matches_dict
        
```

```python
#FUNCION LIGERA, SE MODIFICARA EN UNA FUNCION MAS GENERAL MAS ADELANTE
def generate_data_strike(all_stats,region=region,queue=cola,count=10,save=False,save_name=None):
    data = dict()
    
    for p_id,stats in all_stats.items():
        name = lol_watcher.summoner.by_id(region,p_id)['name']
        x = np.zeros(count+1)
        y = np.zeros(count+1)
        y[0:min(count,len(stats))] += np.array([int(stats[n]['win']) for n in range(min(count,len(stats)))])
        x[0:min(count,len(stats))] += np.array([int(stats[n]['win']) for n in range(min(count,len(stats)))])
        x = np.flip(x)
        y = np.flip(y)
        data[name + '_win/lose'] = y
        x[2+count-len(stats):] = 2*x[2+count-len(stats):] - 1
        for i in range(count):
            x[i+1] = x[i+1]+x[i]
        data[name + '_strike'] = x
    if save:
        if isinstance(save_name,str):
            df_name = save_name + '.csv'
        else:
            df_name = 'sinnombre.csv'
        pd.DataFrame(data).to_csv(df_name)
    return pd.DataFrame(data)
```

```python
from operator import itemgetter
import os

def generate_data_by_features(all_stats,features,region=region,queue=cola,count=10,save=False,save_name=None):
    if isinstance(features,str):
        data_k = [features]
    elif isinstance(features,list):
        data_k = features
    
    data_k = [x for x in data_k if x not in ['name_player','id_player','championName','championId']]
    data = {clave:[] for clave in ['name_player','id_player','championName','championId']+data_k}
    
    for p_id,stats in all_stats.items(): 
        for n in range(min(count,len(stats))):
            data['id_player'].append(p_id)
            data['name_player'].append(lol_watcher.summoner.by_id(region,p_id)['name'])
            data['championName'].append(stats[n]['championName'])
            data['championId'].append(stats[n]['championId'])
            for clave in data_k:
                #parche para evitar error
                try:
                    if clave in stats[n]['challenges'].keys():
                        data[clave].append(stats[n]['challenges'].get(clave))
                    elif clave in stats[n].keys():
                        data[clave].append(stats[n].get(clave)) 
                    else:
                        data[clave].append(None)
                except KeyError:
                    if 'challenges' not in stats[n].keys() and clave == 'kda':
                        try:
                            data[clave].append((stats[n]['kills'] + stats[n]['assists'])/stats[n]['deaths']) 
                        except ZeroDivisionError:
                            if stats[n]['kills'] + stats[n]['assists'] == 0:
                                data[clave].append(0);
                            else:
                                data[clave].append(stats[n]['kills'] + stats[n]['assists'])
                    elif 'challenges' not in stats[n].keys() and clave != 'kda':
                        if clave in stats[n].keys():
                            data[clave].append(stats[n].get(clave))
                        else:
                            data[clave].append(np.nan)
    #print([(clave,len(data[clave])) for clave in data.keys()])
    #SI SE ACTIVA EL PRINT, ES PORQUE HAY UN BUG
    try:
        data_df = pd.DataFrame(data).dropna(axis=1,how='all')
    except:
        print("Problemas al transformar a Dataframe, revisa el codigo. Se ha regresado un diccionario")
        return data
    if save:
        if isinstance(save_name,str):
            df_name = save_name + '.csv'
        else:
            df_name = 'sinnombre.csv'
        path = "C:\\Users\\ikeem\\OneDrive\\Escritorio\\Random-DataScience\\Proyecto1_lol\\Data\\data_by_feat"
        data_df.to_csv(os.path.join(path,df_name),index = False)
    return data_df
```

```python
def plot_winning_strike_teammates(all_stats,count=10,team = False,desacople=False):
    all_names = []
    if desacople:
        fig, axs = plt.subplots(5,2,figsize=(22,17))
        cont = 0;
    for p_id,stats in all_stats.items():
        name = lol_watcher.summoner.by_id(region,p_id)['name']
        all_names.append(name)
        x = np.zeros(count+1)
        x[0:min(count,len(stats))] += np.array([int(stats[n]['win']) for n in range(min(count,len(stats)))])
        if x[0] == 1:
            wincol = True
        elif x[0] == 0:
            wincol = False
        x = np.flip(x)
        x[2+count-len(stats):] = 2*x[2+count-len(stats):] - 1
        for i in range(count):
            x[i+1] = x[i+1]+x[i]
        
        if desacople:
            ax = axs[cont%5,int(cont/5)]
            if team == False:
                ax.plot(x,'o--')
                ax.set_title(name)
            elif wincol == True:
                ax.plot(x,'o--',color='blue')
                ax.set_title(name)
            elif wincol == False:
                ax.plot(x,'o--',color='red')
                ax.set_title(name)
            cont+=1
        else:
            if team == False:
                plt.plot(x,'o--',label=name)
                plt.legend(loc='lower left')
            elif wincol == True:
                plt.plot(x,'o--',color='blue',label=name)
                plt.legend(loc='lower left')
            elif wincol == False:
                plt.plot(x,'o--',color='red',label=name)
                plt.legend(loc='lower left')    
        
    plt.tight_layout()    
    plt.show()
    return None
def plot_winning_strike_player(name_sum,all_stats,count=10):
    for p_id,stats in all_stats.items():
        if lol_watcher.summoner.by_name(region,name_sum)['id'] == p_id:
            name = lol_watcher.summoner.by_id(region,p_id)['name']
            x = np.zeros(count+1)
            x[0:min(count,len(stats))] += np.array([int(stats[n]['win']) for n in range(min(count,len(stats)))])
            x = np.flip(x)
            x[2+count-len(stats):] = 2*x[2+count-len(stats):] - 1
            for i in range(count):
                x[i+1] = x[i+1]+x[i]
            plt.plot(x,'o--',label=name)
            plt.legend(loc='lower left')
    return None
```
