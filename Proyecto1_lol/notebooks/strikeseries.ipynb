```python
#Realizaremos una función que simule una trayectoria para una caminata aleatoria simple, los parámetros a considerar son los siguientes:
#n=número de pasos
#u=posición inicial
#p=probabilidad de avanzar uno en cada paso.

caminata_simple=function(n,u,p){
  r=rbinom(n,1,p)
  e=2*r-1
  e=c(0,e)
  x=0:n
  y=rep(u,n+1)
  for(i in 1:n)
  {y[i+1]=y[i]+e[i+1]}
  plot(x,y,type='o',pch=16,cex=1.2,
       main='Caminata Aleatoria Simple',
       ylab='Espacio de Estados  X_n',
       xlab='Espacio Parametral  Z')
  abline(h=0,v=0,col='red')
  (cbind(x,e,y))
    pacf(ts(y))}

##EJEMPLO:
caminata_simple(100,0,.5)
```
